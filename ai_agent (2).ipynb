# 🤖 AI Automation Agent Demo

This notebook demonstrates a simple AI agent that can:
- Generate emails
- Summarize text
- Suggest code

```python
def ai_agent(task):
    if "email" in task.lower():
        return "Hello, I hope you're doing well. Let’s connect soon."

    elif "summary" in task.lower():
        return "This text has been summarized using AI logic."

    elif "code" in task.lower():
        return "def hello(): print('Hello World')"

    else:
        return "Task processed successfully."
```

```python
print("🤖 AI Automation Agent Started")

user_input = input("Enter your task: ")
```

```python
result = ai_agent(user_input)
print("\n🧠 Result:\n")
print(result)
```

```python
test_tasks = [
    "write an email",
    "give summary of report",
    "write code for function",
    "random task"
]

for task in test_tasks:
    print("\nTask:", task)
    print("Output:", ai_agent(task))
```

## 🤖 Advanced Resume Analysis (RAG + AI Simulation)

```python
def analyze_resume(resume, role):
    required_skills = {
        "Data Scientist": ["Python", "SQL", "Machine Learning", "Statistics", "Data Visualization"]
    }
    detected = [skill for skill in required_skills[role] if skill.lower() in resume.lower()]
    missing = [skill for skill in required_skills[role] if skill not in detected]
    feedback = [f"Consider adding {skill} experience." for skill in missing]
    return detected, missing, feedback
```
